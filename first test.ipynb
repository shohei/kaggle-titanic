```python
from sklearn.linear_model import LogisticRegression
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df2 = df.apply(lambda x: pd.factorize(x)[0])
df2[df2<0]=0
df2
```

```python
data = np.array(df2.drop('Survived', axis=1))
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
scaler.fit(data)
data = scaler.transform(data)
target = (np.array(df2.Survived))
print(data)
print(target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,target)
```

```python
X_test.shape[0]/(X_test.shape[0]+X_train.shape[0])
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
test = pd.read_csv('test.csv')
test_pid = pd.DataFrame({'PassengerId':test['PassengerId']})
test_tmp = test.drop('PassengerId', axis=1)
test_tmp = test_tmp.apply(lambda x: pd.factorize(x)[0])
test_tmp[test_tmp<0]=0
test = pd.concat([test_pid,test_tmp],axis=1)
test
```

```python
survived = clf.predict(test).astype(int)
```

```python
result = pd.DataFrame({"PassengerId":test.PassengerId,"Survived":survived})
result.to_csv('submission.csv',index=False)
```
